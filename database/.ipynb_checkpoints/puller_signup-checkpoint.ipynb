### Setup ###

```python
import mysql.connector
import pandas as pd
import hashlib

# Connect
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="40418",
    database="aeras"
)
print("✅ Connected!")

# Buffered cursor avoids unread result issues
cursor = conn.cursor(buffered=True)

# Functions
def encoding(input_string):
    return hashlib.sha256(input_string.encode('utf-8')).hexdigest()

def genID(iD, signup, phone, vehicle):
    return signup[2:4] + phone[4:6] + str(iD).zfill(3) + vehicle[-2:]

# Get next ID
cursor.execute("SELECT COUNT(*) FROM pullers")
iD = cursor.fetchone()[0] + 1
```

```python

# User data
name = "Kabul Karim"
vehicle = "CHA-C-11-5468"
flag = 0
ban = 0
password = "12345"
points = 0
signup = "2025-11-13"
phone = "01313731509"
ride_count=0
```

```python

# Insert
insert_query = """
INSERT INTO pullers
(pullerid, name, vehicle, flag, ban, password, points, sign_up_date, phone,ride_count)
VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s,%s)
"""

data = (
    encoding(genID(iD, signup, phone, vehicle)),
    encoding(name),
    encoding(vehicle),
    flag,
    ban,
    encoding(password),
    points,
    signup,
    encoding(phone),
    ride_count
)

cursor.execute(insert_query, data)
conn.commit()
print(f"✅ Added {cursor.rowcount} puller")

# Read table with pandas
df = pd.read_sql("SELECT * FROM pullers", conn)
print(df)
```
